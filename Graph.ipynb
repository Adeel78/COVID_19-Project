```python

import time
import matplotlib.pyplot as plotter
import time
time.sleep(5)


pieLabels              = 'USA','Spain','Italy','France','Germany','Rest of the world'

 


Share     = [35, 15.5 ,14, 13.3, 12.5,9.7 ]

 

figureObject, axesObject = plotter.subplots()

 

# Draw the pie chart
axesObject.pie(Share,

        labels=pieLabels,

        autopct='%1.2f',

        startangle=90)

 

# Aspect ratio - equal means pie is a circle

axesObject.axis('equal')

a=plotter.show()
```
